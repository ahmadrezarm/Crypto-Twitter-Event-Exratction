```python
from sklearn.model_selection import train_test_split

from transformers import pipeline
# we translate from english to french, and then from french to english

enTofr = pipeline("translation", model = "Helsinki-NLP/opus-mt-en-fr", device = 0)
frToen = pipeline("translation", model = "Helsinki-NLP/opus-mt-fr-en", device = 0)
```

```python
df = pd.read_excel("/content/A combined v2 (1).xlsx")
df
```

```python
df["lable"] = np.where(df["lable"]=="Markting", "Marketing", df["lable"])
df["lable"] = np.where(df["lable"]=="marketing", "Marketing", df["lable"])
df["lable"] = np.where(df["lable"]=="other", "Other", df["lable"])
df["lable"] = np.where(df["lable"]=="partnership/Investment", "Partnership/Investment", df["lable"])

df = df[df.lable != "km"]
```

```python
y_lable = list(df["lable"])
unique_lables = np.unique(y_lable)
print(unique_lables)
```

```python
df["lable"].value_counts()
```

```python
df_train, df_test = train_test_split(df, test_size = 0.2, stratify=df[["lable"]])
```

```python
df_train["lable"].value_counts()
```

```python
df_test["lable"].value_counts()
```

```python
df_train.to_excel("training_dataset.xlsx")
df_test.to_excel("test_dataset.xlsx")
```

```python
feature_release_tweets["augmented_tweets"] = feature_release_tweets["tweets"].apply(lambda x: frToen(enTofr(x)[0]["translation_text"])[0]["translation_text"])
feature_release_tweets
```

```python
combined_feature_release_tweets = [feature_release_tweets[["tweets", "lable"]], feature_release_tweets[["augmented_tweets", 'lable']]]
combined_feature_release_tweets= pd.concat(combined_feature_release_tweets)
combined_feature_release_tweets.duplicated().sum()
```
